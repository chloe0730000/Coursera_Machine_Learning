```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline  
```

```python
data = pd.read_csv("places.txt", header = None)
data = data.rename(columns = {0:"longitude", 1: "latitude"})
data = data[["longitude", "latitude"]]

clf = KMeans(n_clusters=3, n_jobs=-1)
data["places"] = clf.fit_predict(data)
```

```python
data
```

```python
data["id"] = data.index
data = data[["id", "places"]]
#final.to_csv("upload_result.csv", encoding = "utf-8-sig", index=False, header=False)
data.to_csv("clusters.txt", encoding = "utf-8-sig", index=False, header=False,mode='a')
```
